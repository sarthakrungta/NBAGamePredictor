```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('archive/games.csv')
```

```python
df.head()
```

```python
df.info()
```

**Observations:** 
 - There are null values in our dataset. 

```python
#CHECKING FOR NULL VALUES
df.isnull().sum()
```

```python
#There are about 100 null values. For better understanding, we can view these rows
df[df.isna().any(axis=1)]
```

**Observations:** 
 - The missing values are from old pre season games from 2003
 - As such, we can drop these rows from our model

```python
#Removing Rows with Null values
df = df.dropna(how='any',axis=0)
```

```python
#Some of the columns in the database seem to be duplicates such as HOME_TEAM_ID, TEAM_ID_home and VISITOR_TEAM_ID, TEAM_ID_away
#Let us confirm this assumption and if so, we can drop duplicate columns

print("Differences in HOME_TEAM_ID and TEAM_ID_home: {}".format(df[df['HOME_TEAM_ID'] != df['TEAM_ID_home']].shape[0]))
print("Differences in VISITOR_TEAM_ID and TEAM_ID_away: {}".format(df[df['VISITOR_TEAM_ID'] != df['TEAM_ID_away']].shape[0]))
```

```python
#Since all values in the two sets are the same, we can drop duplicate columns
df = df.drop(['TEAM_ID_home','TEAM_ID_away'],axis=1)
```

```python
#We can also see that GAME_STATUS_TEXT is always 'Final'. We can ensure the same and if so, get rid of the redundant column

print("Occurances of GAME_STATUS_TEXT not 'Final': {}".format(df[df['GAME_STATUS_TEXT'] != 'Final'].shape[0]))
```

```python
#Since all values in the column are the same, we can remove the column from our model
df = df.drop('GAME_STATUS_TEXT',axis=1)
```

```python
#We know that HOME_TEAM_ID, VISITOR_TEAM_ID and SEASON are categorical values as there are a limited number of teams and seasons
#Convert these columns from int64 to categorical values

categorical_columns = ['HOME_TEAM_ID','VISITOR_TEAM_ID','SEASON']
for col in categorical_columns:
    df[col] = df[col].astype('category')
    
df.describe(include = 'category')
```

```python
#GAME_DATE_EST needs to also be converted into a dateType column
df['GAME_DATE_EST'] = pd.to_datetime(df['GAME_DATE_EST'])
```

```python
df.info()
```

**Observations:** 
 - There are no Null values in our Data
 - All Features are of the desired Data Type

## EDA

```python
## We will start by exploring each column distribution of the data
plt.figure(figsize=(25,30))
for i,col in enumerate(df.columns):
    plt.subplot(5,5,i+1)
    plt.hist(df[col], bins=20)
    plt.title(col)
```

```python
#Now, we can look at the correlation between different features
f, ax = plt.subplots(figsize=(14, 11))
corr = df.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    annot=True
)
```

**Observations**
-  FG_PCT and PTS are highly correlated with each other
-  AST and PTS are highly correlated with each other
-  FG_PCT_HOME is negatively correlated to REB_AWAY
-  FG_PCT_AWAY is negatively correlated to REB_HOME
